```python
import pandas as pd
```

```python
dipresive = pd.read_csv("anxiety OR apathy OR depression OR PTSD OR thought of suicide.csv")
normal = pd.read_csv("positive OR love OR motivation OR happy OR inspiration.csv")
```

```python
dipresive['target'] = 'Dipresive'
normal['target'] = 'Normal'
```

```python
df = df.drop_duplicates(subset=['tweet'], keep = False)
```
