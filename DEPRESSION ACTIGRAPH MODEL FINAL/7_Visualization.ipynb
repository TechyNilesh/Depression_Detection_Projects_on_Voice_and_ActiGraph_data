```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("./depresjon-dataset/condition/condition_1.csv")
df2 = pd.read_csv("./depresjon-dataset/control/control_1.csv")
```

```python
df1 = df1.rename(columns={"activity": "Activity_conditionl"})
df2 = df2.rename(columns={"activity": "Activity_Control"})
```

```python
df1 = df1.drop(['timestamp','date'],axis=1)
df2 = df2.drop(['timestamp','date'],axis=1)
```

```python
df = pd.concat([df1,df2], axis=1)
df = df.head(60)
```

```python
df.plot()
plt.title('Total 1-Hours Activity Graph For [Control] & [Condition] Group')
plt.xlabel('Time (Min)')
plt.ylabel('Activity')
```

```python
df1 = pd.read_csv("./depresjon-dataset/condition/condition_1.csv", parse_dates=["timestamp"], index_col="timestamp")
df2 = pd.read_csv("./depresjon-dataset/control/control_1.csv",parse_dates=["timestamp"], index_col="timestamp")
```

```python
df1 = df1.rename(columns={"activity": "Activity_conditionl"})
df2 = df2.rename(columns={"activity": "Activity_Control"})
```

```python
df1 = df1['Activity_conditionl'].resample('60T').mean()
df2 = df2['Activity_Control'].resample('60T').mean()
```

```python
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
```

```python
df1 = df1.reset_index()
df2 = df2.reset_index()
```

```python
df1 = df1.drop(['timestamp'],axis=1)
df2 = df2.drop(['timestamp'],axis=1)
```

```python
df = pd.concat([df1,df2], axis=1)
df = df.head(24)
```

```python
df.plot()
plt.title('Total 1-Day Activity Graph For [Control] & [Condition] Group')
plt.xlabel('Time (hr)')
plt.ylabel('Activity')
```
